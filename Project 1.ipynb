```python
# Question 1
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
df = pd.read_csv('Employee_Payroll.csv')
df = df.dropna()  
df_new = df[['Fiscal Quarter','Office', 'Job Title','Base Pay']]
df_new1 = df_new[df['Job Title']=='Director']
df_new2 = df_new[df['Job Title']=='Sheriff']
df_new3 = df_new[df['Job Title']=='Post Graduate Level Physician']
df_new1 = df_new1[df['Fiscal Year']==2016]
df_new2 = df_new2[df['Fiscal Year']==2016]
df_new3 = df_new3[df['Fiscal Year']==2016]



df = df.dropna()
print(df_new1.shape)
print(df_new2.shape)
print(df_new3.shape)

# print(df.head(100))
# print (df_new.head(100)) 



```

```python
# Question 1
df_new1 = df_new1.drop_duplicates()
print (df_new1)

df_new2 = df_new2.drop_duplicates()
print (df_new2)

df_new3 = df_new3.drop_duplicates()
print (df_new3)
```

This data source is the employee payroll in cook county illinos. My main question is out of the three jobs that I have chosen to investigate, what job in illinois make the most money. Im looking at data from cook county illinos Cook County is the most populous county in the U.S. state of Illinois and the second most populous county in the United States More than 40% of all residents of illinois live within Cook county. My first, second and third questions are what is the highest salary of the three jobs, What is the average salary of the three jobs, and What is the lowest salary of the three jobs. This will tell me the amount of money within that field that residents in this county are making from the lowest to the highest salary. I am basing this research on the year of 2016 so im only taking data from the year. Also the three jobs that I am choosing to look into are Director, Sheriff, and Post Graduate Level Physician. The county fiscal quarters are: Q1: December - February Q2: March - May Q3: June - August Q4: September - November.

```python
# Prints the highest salary of each of the jobs.
plt.bar(df_new1['Job Title'],df_new1['Base Pay']) 
plt.bar(df_new2['Job Title'],df_new2['Base Pay']) 
plt.bar(df_new3['Job Title'],df_new3['Base Pay'])
plt.title('Cook County Salary')
plt.xlabel('Job Titles ')
plt.ylabel('Salary')
plt.legend(['Second Lowest','Highest','Lowest'], loc = "upper right")

plt.show()

```

I have discovered that the Sheriff position makes the most money out of the three jobs that I chose and that the Post Graduate Level Physician makes the least amount of money in Cook county in 2016.

```python
# Prints and plot the max base pay values of the three jobs in the Cook county salary data set
print (round(df_new1['Base Pay'].max()))
print (round(df_new2['Base Pay'].max()))
print (round(df_new3['Base Pay'].max()))
plt.bar(df_new1['Job Title'],df_new1['Base Pay'].max()) 
plt.bar(df_new2['Job Title'],df_new2['Base Pay'].max()) 
plt.bar(df_new3['Job Title'],df_new3['Base Pay'].max())
plt.title('Cook County Salary')
plt.xlabel('Job Titles ')
plt.ylabel('Salary')


plt.show()
```

```python
# calculate and print the mean of the base salary and plot the average mean of the cook salary
print (round(df_new1['Base Pay'].mean()))
print (round(df_new2['Base Pay'].mean()))
print (round(df_new3['Base Pay'].mean()))

plt.barh(df_new1['Job Title'],df_new1['Base Pay'].mean() )
plt.barh(df_new2['Job Title'],df_new2['Base Pay'].mean() )
plt.barh(df_new3['Job Title'],df_new3['Base Pay'].mean() )

plt.title('Cook County Salary')
plt.xlabel('Salary ')
plt.ylabel('Job Titles')


plt.legend(['Second Lowest','Highest','Lowest'], loc = "upper right")


plt.show()
```
